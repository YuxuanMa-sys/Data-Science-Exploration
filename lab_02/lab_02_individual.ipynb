### <u>Tutorial</u>: Other Aggregation Functions

In addition to the **.min()**, **.max()**, and **.sum()** aggregation functions, we can also use the **.mean()** function to calculate the mean and the **.median()** function to calculate the median of a pandas series.



````python
tmp_df['a'].mean()
````

````python
tmp_df['a'].median()
````

### <u>Tutorial</u>: Filtering for Multiple Conditions

We can filter a dataframe using multiple conditions using the **&** operator and the **|** operator.



The **&** operator is an **'and'** operator that stipulates that both conditions on either side of it must be met. You must put your conditions in parantheses if you plan on stipulating more than one.

````python
tmp_df[(tmp_df['a']>2) & (tmp_df['a']<=4)]
````

````python
tmp_df[(tmp_df['a']>2) & (tmp_df['a']<=4) & (tmp_df['b']=='x')]
````

The **|** operator is an **'or'** operator that stipulates that either conditions on either side of it could be met.

````python
tmp_df[(tmp_df['a']>2) | (tmp_df['b']=='x')]
````

````python
tmp_df[(tmp_df['a']>2) | (tmp_df['b']=='x') | (tmp_df['b']=='y')]
````
